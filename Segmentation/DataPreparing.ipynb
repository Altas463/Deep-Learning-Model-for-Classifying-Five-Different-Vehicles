```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import os
```

```python
# Đường dẫn đến thư mục chứa ảnh
image_folder_path = "/content/drive/MyDrive/Deep/Segment_Data/Images/Bicycle"
label_folder_path = "/content/drive/MyDrive/Deep/Segment_Data/Labels/Bicycle"

# Lấy danh sách tên tệp ảnh trong thư mục và sắp xếp theo thứ tự số
image_file_names = natsorted(os.listdir(image_folder_path))

bicycle_images = []
bicycle_labels = []

# Duyệt qua từng tệp ảnh trong thư mục
for image_filename in image_file_names:
    # Kiểm tra định dạng của tệp ảnh (png/jpg)
    if image_filename.endswith(".png") or image_filename.endswith(".jpg"):
        # Đường dẫn đầy đủ đến tệp ảnh
        image_path = os.path.join(image_folder_path, image_filename)

        # Mở ảnh bằng OpenCV
        img = cv2.imread(image_path)
        bicycle_images.append(img)

        # Tìm tên tệp ảnh tương ứng trong thư mục nhãn
        label_filename = None
        for label_file in os.listdir(label_folder_path):
            if label_file.startswith(image_filename[:-4]):
                label_filename = label_file
                break

        # Kiểm tra nếu tìm thấy tệp nhãn tương ứng
        if label_filename:
            label_path = os.path.join(label_folder_path, label_filename)
            # Thực hiện xử lý nhãn tại đây
            label_img = cv2.imread(label_path,cv2.IMREAD_GRAYSCALE)
            bicycle_labels.append(label_img)

        # Hiển thị thông tin ảnh và nhãn
        print("\n")
        print("Tên tệp ảnh:", image_filename)
        print("Tên tệp nhãn:", label_filename)

        # Hiển thị ảnh
        cv2_imshow(img)
        cv2_imshow(label_img)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
print("\n Hoàn tất quá trình")
bicycle_labels = np.array(bicycle_labels)
bicycle_labels[bicycle_labels >0] = 1
bicycle_images = np.array(bicycle_images)
```

```python
print('Size of Bicycle Images:',bicycle_images.shape)
print('Size of Bicycle Labels:',bicycle_labels.shape)
print('Unique value:',np.unique(bicycle_labels))
```
