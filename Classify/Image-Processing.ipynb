```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.svm import SVC
import cv2
import os
from google.colab import files
import glob
```

```python
# Đường dẫn đến thư mục chứa ảnh trên máy tính
folder_path = '/content/drive/MyDrive/Deep/Deep-Data/Data/XeTai'

# Đường dẫn đến thư mục lưu trữ ảnh đã resize
output_folder = '/content/drive/MyDrive/Deep/Deep-Data/Data/Truck'

# Kích thước mong muốn
target_size = (224, 224)

# Tìm kiếm tất cả các tệp tin ảnh trong thư mục
image_files = glob.glob(os.path.join(folder_path, '*.jpg')) + glob.glob(os.path.join(folder_path, '*.png'))

# Lặp qua tất cả các tệp tin ảnh
for image_file in image_files:
    # Đường dẫn đến tệp tin
    image_path = os.path.join(folder_path, image_file)

    # Đọc ảnh
    image = cv2.imread(image_path)

    # Resize ảnh xuống kích thước lớn nhất có thể (theo chiều rộng hoặc chiều cao)
    max_dimension = max(image.shape[0], image.shape[1])
    if max_dimension > target_size[0] or max_dimension > target_size[1]:
        scale_factor = min(target_size[0] / max_dimension, target_size[1] / max_dimension)
        new_width = int(image.shape[1] * scale_factor)
        new_height = int(image.shape[0] * scale_factor)
        resized_image = cv2.resize(image, (new_width, new_height))
    else:
        resized_image = image

    # Padding ảnh để có kích thước 224x224
    padded_image = np.zeros((target_size[0], target_size[1], 3), dtype=np.uint8)
    start_x = (target_size[1] - resized_image.shape[1]) // 2
    start_y = (target_size[0] - resized_image.shape[0]) // 2
    padded_image[start_y:start_y+resized_image.shape[0], start_x:start_x+resized_image.shape[1]] = resized_image

    # Tạo thư mục đầu ra nếu chưa tồn tại
    os.makedirs(output_folder, exist_ok=True)

    # Đường dẫn đến tệp tin ảnh đã resize
    output_path = os.path.join(output_folder, os.path.basename(image_file))

    # Lưu ảnh đã resize
    cv2.imwrite(output_path, padded_image)
    print("Hoàn thành quá trình resize ảnh.")
```
